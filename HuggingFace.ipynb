### Load Pretrained Model

```python
#model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
preprocessor = EfficientNetImageProcessor.from_pretrained("dennisjooo/Birds-Classifier-EfficientNetB2")
model = EfficientNetForImageClassification.from_pretrained("dennisjooo/Birds-Classifier-EfficientNetB2")
```

### Process the Image

```python
def huggingface(img):
    image=Image.open(img)
    inputs = preprocessor(images=image, return_tensors="pt")
    outputs=model(**inputs)
    logits=outputs.logits
    predicted_class_idx=logits.argmax(-1).item()
    #print("Predicted class:", model.config.id2label[predicted_class_idx])
    return model.config.id2label[predicted_class_idx]
```

## Prediction

```python
img1=huggingface(img1)
img2=huggingface(img2)
img3=huggingface(img3)
img4=huggingface(img4)
img5=huggingface(img5)
img6=huggingface(img6)
print(f"The predicted class for img1 is {img1}")
print(f"The predicted class for img2 is {img2}")
print(f"The predicted class for img3 is {img3}")
print(f"The predicted class for img4 is {img4}")
print(f"The predicted class for img5 is {img5}")
print(f"The predicted class for img6 is {img6}")
```
